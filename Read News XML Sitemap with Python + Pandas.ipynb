```python
import requests
import pandas as pd
import xmltodict

url = "https://www.nzz.ch/sitemap/news0.xml	"
res = requests.get(url)
raw = xmltodict.parse(res.text)

data = [[r["loc"], r["news:news"]["news:title"], r["news:news"]["news:publication_date"]] for r in raw["urlset"]["url"]]
df = pd.DataFrame(data, columns=["loc","news:title", "news:publication_date"])
df
```
